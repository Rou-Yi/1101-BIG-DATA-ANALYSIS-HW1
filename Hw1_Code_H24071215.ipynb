### 資料前處理

```python
# 區分連續與量化資料
def split_data_by_feature_types(df):
    df_categorial, df_quantitative = [], []
    for col_name, types in zip(df.columns, df.dtypes):
        if types in ['object','bool']:
            df_categorial.append(col_name)
        else:
            df_quantitative.append(col_name)
    df_categorial = df[df_categorial]
    df_quantitative = df[df_quantitative]
    
    return df_categorial, df_quantitative
```

```python
def data_pre_processing(data_X):
    # 區分連續與量化資料
    categorial_df, quantitative_df = split_data_by_feature_types(data_X)
    
    cat_cols   = categorial_df.columns  #類別變數
    num_cols   = quantitative_df.columns  #數值變數
    bin_cols   = data_X.nunique()[data_X.nunique() == 2].keys().tolist()  #二元變數
    multi_cols = [i for i in cat_cols if i not in bin_cols]  #多元變數
    
    # 過濾多於 100 個種類的類別變數
    for name in cat_cols:
        if categorial_df[name].nunique()>100:
            del categorial_df[name]
            cat_cols = cat_cols.drop(name)
    
    #對二元變數編碼
    encoder = LabelEncoder()
    for i in bin_cols :
        data_X[i] = encoder.fit_transform(data_X[i])

    #對多元變數編碼
    data_X = pd.get_dummies(data = data_X, columns = multi_cols)

    #標準化數值變數
    std = StandardScaler()
    data_X[num_cols] = std.fit_transform(data_X[num_cols])
    
    return data_X
```

```python
# 資料分割
from sklearn.model_selection import train_test_split
# 模型評估
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, auc
from statistics import mean
```

```python
config = {'trials': 5,
          'train_ratio': 0.65,
          'val_ratio': 0.10,
          'test_ratio': 0.25,
          'random_state':[56, 1215, 6400, 508, 21]
}
```

```python
## a general definition to predict validation data with any model
def model_prediction(model, X_train_data, y_train_data):
    total_auc, total_accuracy = [], []
    trials = config['trials']
    for i in range(trials):
        # splitting train and validation data with 5 random seeds
        X_train, X_val, y_train, y_val = train_test_split(X_train_data, y_train_data, 
                                               test_size=config['val_ratio']/(config['val_ratio']+config['train_ratio']), 
                                               random_state=config['random_state'][i], shuffle=True)

        # Learn the digits on the train subset
        model.fit(X_train, y_train)
        # Predict the value of the digit on the validation subset
        y_pred = model.predict(X_val)
        
        y_pred_df = pd.DataFrame({"pred":y_pred})
        pred_pos_persent = y_pred_df.value_counts()[1]/len(y_pred_df)
        accuracy = accuracy_score(y_val, y_pred)
        total_accuracy.append(accuracy)
        auc = roc_auc_score(y_val, model.predict_proba(X_val)[:,1])
        total_auc.append(auc)
    
    print("Mean auc : {:.2%}  |  Mean accuracy : {:.2%}".format(mean(total_auc), 
                                                                mean(total_accuracy)))
    return mean(total_auc), mean(total_accuracy)
```

```python
# 打亂資料
from sklearn.utils import shuffle

## a general definition to predict validation data with any model
## using smote
def model_prediction_unsupervised(model, X_train_data, y_train_data):
    total_auc, total_accuracy = [], []
    trials = config['trials']
    for i in range(trials):
        # take 50 of data as train data
        shuffle_X = shuffle(X_train_data, random_state=config['random_state'][i])
        shuffle_y = shuffle(y_train_data, random_state=config['random_state'][i])
        X_train = shuffle_X.iloc[:50, :]
        y_train = shuffle_y.iloc[:50]
        X_test = shuffle_X.iloc[50:, :]
        y_test = shuffle_y.iloc[50:]
        
        # Learn the digits on the train subset
        model.fit(X_train, y_train)
        # Predict the value of the digit on the validation subset
        y_pred = model.predict(X_test)
        
        accuracy = accuracy_score(y_test, y_pred)
        total_accuracy.append(accuracy)
        auc = roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
        total_auc.append(auc)
    
    print("Mean auc : {:.2%}  |  Mean accuracy : {:.2%}".format(mean(total_auc), 
                                                                mean(total_accuracy)))
    return mean(total_auc), mean(total_accuracy)
```

```python
# 資料平衡
from imblearn.over_sampling import SMOTE

## a general definition to predict validation data with any model
## using smote
def model_prediction_with_smote(model, X_train_data, y_train_data):
    total_auc, total_accuracy = [], []
    trials = config['trials']
    for i in range(trials):
        # oversampling minority class using smote against the train data
        X_os_smote, y_os_smote = SMOTE(random_state = 0).fit_resample(X_train_data, y_train_data)

        # splitting train and validation data with 5 random seeds
        X_train, X_val, y_train, y_val = train_test_split(X_os_smote, y_os_smote,
                                               test_size=config['val_ratio']/(config['val_ratio']+config['train_ratio']), 
                                               random_state=config['random_state'][i], shuffle=True)

        # Learn the digits on the train subset
        model.fit(X_train, y_train)
        # Predict the value of the digit on the validation subset
        y_pred = model.predict(X_val)
        
        y_pred_df = pd.DataFrame({"pred":y_pred})
        pred_pos_persent = y_pred_df.value_counts()[1]/len(y_pred_df)
        accuracy = accuracy_score(y_val, y_pred)
        total_accuracy.append(accuracy)
        auc = roc_auc_score(y_val, model.predict_proba(X_val)[:,1])
        total_auc.append(auc)
        
    print("Mean auc : {:.2%}  |  Mean accuracy : {:.2%}".format(mean(total_auc), 
                                                                mean(total_accuracy)))
    return mean(total_auc), mean(total_accuracy)
```
